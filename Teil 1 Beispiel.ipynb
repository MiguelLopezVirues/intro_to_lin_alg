# Total realistisches Praxisbeispiel: Zombie-Apokalypse in Deiner Analytics-Abteilung

## Setting

Nehmen wir mal an, Deine Analytics-Abteilung besteht aus 100 Personen. Am ersten Tag ist eine Person ist ein Zombie und 99 Personen sind wohl auf (gesund).

```python
z = 1
g = 99
```

Basteln wir mal einen Vektor daraus, nennen ihn `at_erster_tag` und geben ihn aus. Als Datentyp verwenden wir hier `ndarray`. Es gibt auch andere Datentypen, die wir hier nutzen k√∂nnten - z.B. `matrix` oder `vector`. Aber wie eben gesagt ist der Datentyp `ndarray` ist f√ºr's Erste am vielseitigsten. 

```python
at_erster_tag = np.array([z, g])
print(at_erster_tag)
```

Zombies sind nat√ºrlich eine fiese Sache. Gehen wir mal davon aus, dass eine gesunde Person mit 80%-iger Wahrscheinlichkeit an einem Tag von einem Zombie gebissen wird und dann mit 90%-iger Wahrscheinlichkeit f√ºr einen Tag ein Zombie bleibt. Es ergibt sich also die folgende Matrix.

| |heute Zombie|heute gesund|
|----|----|----|
|am n√§chsten Tag Zombie|0.9|0.2|
|am n√§chsten Tag gesund|0.1|0.8|

Setzen wir das ganze mal in Python um:

```python
# Wahrscheinlichkeiten pro Tag

zz = 0.9 # Zombie bleiben
zg = 0.1 # Zombie erholt sich und wird gesund

gg = 0.8 # Gesunde Person bleibt gesund
gz = 0.2 # Gesunde Person wird von Zombie gebissen

#                        heute Zombie  heute gesund
#                                  vv  vv
zombifizierungmatrix =  np.array([[zz, gz],  # <- am n√§chsten Tag Zombie
                                  [zg, gg]]) # <- am n√§chsten Tag gesund
print(zombifizierungmatrix)
```

Wir haben die Wahrscheinlichkeiten nun eine Matrix √ºbertragen und sie `zombifizierungsmatrix` genannt. Vielleicht ist Dir aufgefallen, dass die Summe der Wahrscheinlichkeiten in einer Spalte jeweils 1 ergibt. Das ist eine wichtige Eigenschaft dieser Matrix, auch unter dem Namen [√úbergangsmatrix](https://de.wikipedia.org/wiki/√úbergangsmatrix) oder auch Markov-Matrix firmiert. Das bedeutet n√§mlich dass der Raum der M√∂glichkeiten vollst√§ndig ist: Entweder ist man Zombie oder nicht. Eine andere M√∂glichkeit gibt es hier nicht.

## The Zu-Fu√ü-Way
Dann rechnen wir mal zu Fu√ü! Wir nehmen uns die Variablen oben und und multipizieren und addieren uns die Zahlen f√ºr den n√§chsten Tag zusammen.

```python
z_zweiter_tag = z * zz + g * gz # Die Zombies (z), die welche bleiben (zz) plus die Gesunden (g), die Zombies werden (gz).
g_zweiter_tag = g * gg + z * zg # Die Gesunden (g), die welche bleiben (gg) plus die Zombies (z), die Gesund werden (zg).
print(f"Es gibt am 2. Tag {z_zweiter_tag:2.4} Zombies und {g_zweiter_tag:2.4} Gesunde. Es gibt {(z_zweiter_tag/g_zweiter_tag):2.4} mal mehr Zombies als Gesunde.")
```

Okay, das ging ja noch! Jetzt m√ºssen wir noch zwei Probleme tacklen:

1. Was, wenn wir 500 Artikel und nicht 2 Gruppen haben? Das wird nicht funktionieren! Bedeutet: Das Vorgehen **skaliert nicht mit der Vektorgr√∂√üe**.
2. Wie sieht es denn nach 20 Tagen aus? Bedeutet: Das Vorgehen **skaliert nicht im Zeitverlauf**.



## The Lineare-Algebra-Way
Wenn Du Dir die Rechnung f√ºr die Zombies am n√§chsten Tag oben anschaust, f√§llt Dir vielleicht auf, dass Dir die Formel schon einmal untergekommen ist. Es ist das Punktprodukt zwischen zwei Vektoren: N√§mlich dem Vektor `at_erster_tag` mit der ersten Zeile der Matrix `zombifizierungmatrix[0,:] = [0.9 0.2]` . (In Excel nennt sich die Formel √ºbrigens `SUMMENPRODUKT()`). Probieren wir das doch mal aus! Das Punktprodukt wird bei dem Datentyp `ndarray` mit dem Zeichen `@` ausgedr√ºckt.

```python
zombie_quoten = zombifizierungmatrix[0,:]
print("Zombie-Quoten: " + str(zombie_quoten))
gesund_quoten = zombifizierungmatrix[1,:]
print("Gesund-Quoten: " + str(gesund_quoten))
```

```python
z_zweiter_tag = at_erster_tag @ zombie_quoten
g_zweiter_tag = at_erster_tag @ gesund_quoten
print(f"Es gibt am 2. Tag {z_zweiter_tag:2.4} Zombies und {g_zweiter_tag:2.4} Gesunde. Es gibt {(z_zweiter_tag/g_zweiter_tag):2.4} mal mehr Zombies als Gesunde.")
```

Hmmmmmm... So richtig was gewonnen haben wir jetzt noch nicht, oder? Wir haben jetzt etwas eleganteren Code und haben uns gespart, die Spalten der Matrix durchgehen zum m√ºssen. Aber bei Problem 1 und 2 ist uns noch nicht viel geholfen. √úberlegen wir erstmal zu Problem 1: W√§re es nicht cool, wenn man das nicht f√ºr jede Komponente des Vektors einzeln machen m√ºsste? Auch das geht mit dem Zeichen `@`. 

### Die Matrixmultiplikation

Das `@` bedeutet n√§mlich **Matrixmultiplikation**. Das Punktprodukt ist nur ein Spezialfall, wenn wenn beide Matizen nur eine Spalte oder eine Zeile haben. Wir k√∂nnen jetzt einfach den Vektor mit der Matrix multiplizieren und *zack* - das Ergebnis ist da! 

```python
at_zweiter_tag = zombifizierungmatrix @ at_erster_tag
print(f"Es gibt am 2. Tag {at_zweiter_tag[0]:2.4} Zombies und {at_zweiter_tag[1]:2.4} Gesunde. Es gibt {(at_zweiter_tag[0]/at_zweiter_tag[1]):2.4} mal mehr Zombies als Gesunde.")
```

In Excel nennt sich die Formel √ºbrigens `MMULT()`.

![mmult](pics/xls_mmult.PNG)

Wir haben jetzt "nur" eine Zeile Code gespart. Aber bei 500 Artikeln h√§tten wir 499 Zeilen Code gespart. Das ist doch was! Wir haben jetzt eine - Achtung! - Operation gefunden, die immer funktioniert - egal wie lang der Vektor ist. Das ist besonders praktisch, weil wir zur Laufzeit ja h√§ufig gar nicht wissen, mit wie vielen Artikeln oder Kunden oder M√§rkten oder whatever wir es zu tun haben. 

Zur√ºck zur Theorie und speaking of **"Operation"**: Du erinnerst Dich? Ganz am Anfang habe ich geschrieben, dass sich die Algebra mit Rechenoperationen und deren Eigenschaften besch√§ftigt. Ich hoffe, das ist jetzt schon weniger abstrakt geworden. (Auch wenn wir uns die Eigenschaften, gleich noch einmal genauer angucken m√ºssen.) Ich habe auch geschrieben, dass nicht immer Zahlen Gegenstand einer Operation sein m√ºssen. In unserem Fall haben wir eine Abteilung am ersten Tag mit einer Zombifizierung "multipliziert" und die Abteilung am zweiten Tag erhalten. Verr√ºckt, oder? Oder anders ausgedr√ºckt: "matrices do things to vectors". Genau das ist die Magic der Linearen Algebra! :) 

Die Matrixmultiplikation ist wirklich eine ganz tolle Operation. Sie ist so vielf√§ltig, dass man fast keine anderen Operationen ben√∂tigt. Sie kann z.B. Vektoren verl√§ngern und drehen. Sie kann prognostizieren, wie die Welt am n√§chsten Tag aussieht - sowohl beim Wetter als auch bei Zombies. Sie kann Zeilen oder Spalten von anderen Matrizen vertauschen, addieren und subtrahieren. Sie kann Gleichungssysteme l√∂sen und so vieles mehr! Aber Eines nach dem Anderen...

Uff! Problem 1 w√§re also gel√∂st! Jetzt zu Problem 2: Wie sieht es denn am 20. Tag aus? Daf√ºr gibt es drei L√∂sungen:
1. **Der Controller**: Einmal (im Taschenrechner) multiplizieren und dann 20 mal auf die `=`-Taste kloppen.
2. **Der Informatiker**: Eine `for`-Schleife drum schn√ºren und gut is.
3. **Der Ingenieur**: even more Lineare Algebra! <3
4. (**Der Mathematiker**: K√∂nnte uns beweisen, warum das alles so ist.)

Okay, starten wir mit dem **Controller**. Die Idee w√§re jetzt, die Multiplikation mehrfach auszuf√ºhren. Zun√§chst definieren wir den ersten Tag als "heute" und gehen dann in der Zeit immer weiter in die Zukunft.

```python
at_heute = at_erster_tag
```

F√ºhre die folgende Zelle mehrfach mit `Strg-Enter` aus. Wenn Du zur√ºck zum Start m√∂chtest, f√ºhre die obige Zelle noch einmal aus. F√§llt Dir schon etwas auf? Wenn ja, merk's Dir! Wir kommen gleich darauf zur√ºck.

```python
at_morgen = zombifizierungmatrix @ at_heute # Hier berechnen wie den Zombie-Count von morgen
print(f"Es gibt {at_morgen[0]:2.4} Zombies und {at_morgen[1]:2.4} Gesunde. Es gibt {(at_morgen[0]/at_morgen[1]):2.4} mal mehr Zombies als Gesunde.") # Wir geben den Count aus
at_heute = at_morgen # Wir gehen eine Tag in die Zukunft indem wir heute = morgen setzen. #TomorrowIsTheNewToday 
```

Jetzt zum **Informatiker**: 20 mal Enter dr√ºcken, kann man doch auch automatisieren, oder? Los geht's! Dabei hilft uns die zweitbekannteste Schleife der Welt: Die `for`-Loop. Und weil wir "richtige" Informatiker sind machen direkt eine Funktion draus:

```python
def zombifizierung(start_werte, z_matrix, days: int, verbose = False):
    at_heute = start_werte
    ergebnis_historie = []

    for i in range(days) :
        at_morgen = z_matrix @ at_heute # Hier berechnen wie den Zombie-Count von morgen.
        if verbose: # Hier geht es nur darum, dass wir bei Bedarf die Zwischenergebnisse anzeigen lassen k√∂nnen.
            print(f"Es gibt am {i+1}. Tag {at_morgen[0]:2.4} Zombies und {at_morgen[1]:2.4} Gesunde. Es gibt {(at_morgen[0]/at_morgen[1]):2.4} mal mehr Zombies als Gesunde.") # Wir geben den Count aus
        ergebnis_historie.append(at_morgen) # Historie wegschreiben
        at_heute = at_morgen # Wir gehen einen Tag in die Zukunft indem wir heute = morgen setzen. #TomorrowIsTheNewToday 

    return np.array(ergebnis_historie)
```

Jetzt rufen wir die Funktion auf
- Wir geben die Startbedingung mit: 1 Gesunde, 99 Zombies
- Die Ver√§nderungsraten: Zombies bleiben zu 90% Zombies etc.
- Wie viele Tage wir in die Zukunft gucken m√∂chten: 20
- und, ob wir die Zwischenergebnisse ausgeben m√∂chten: Yes!

```python

ergebnis_historie = zombifizierung(at_erster_tag, zombifizierungmatrix, 20, True)
```

#### Eigenwerte
Du hast wahrscheinlich weiter oben schon gesehen, dass nicht nur Eigenvektoren ausgegeben wurden sondern auch Eigenwerte. Das Wort hast Du wahrscheinlich schon einmal bei einer PCA oder Faktorenanalyse geh√∂rt. Schauen wir uns noch einm mal genau an, was drin steht:


```python
eigvals, eigvectors = LA.eig(zombifizierungmatrix)
evec1 = eigvectors[:,0]
print(f"Der Eigenvektor ist: {evec1}" )
print(f"Die L√§nge des Eigenvektors ist: {LA.norm(evec1, 2)}")
print(f"Der Eigenwert ist: {eigvals[0]}")
print(f"Es gibt langfristig {evec1[0]/evec1[1]:2.4} mal mehr Zombies als Gesunde.")
```

Okay, da kommt "1" raus. Was mag das wohl bedeuten? Also an den Startbedingungen kann es ja nicht liegen, denn das Ergebnis der `Eig-`Methode kann ja nur etwas mit der Matrix zu tun haben. Wagen wir doch mal ein Experiment: Was passiert, wenn wir die Matrix √§ndern? Mal angenommen, die Gesundheitsrate und die Zombiefizierung halten sich mit jeweils 20% die Waage:

```python
zombifizierungmatrix_symm =  np.array([[0.8, 0.2],  # <- am n√§chsten Tag Zombie
                                       [0.2, 0.8]]) # <- am n√§chsten Tag gesund
print(zombifizierungmatrix_symm)

(eigvals_symm, eigvectors_symm) = LA.eig(zombifizierungmatrix_symm)
evec1_symm = eigvectors_symm[:, 0]
eval1_symm = eigvals_symm[0]
print(f"Der Eigenvektor ist: {evec1_symm}" )
print(f"Die L√§nge des normierten Eigenvektors ist: {LA.norm(evec1_symm, 2)}")
print(f"Die L√§nge des Eigenvektors ist: {LA.norm(evec1_symm*eval1_symm, 2)}")
print(f"Der Eigenwert ist: {eval1_symm:2.4}")
print(f"Es gibt langfristig {evec1_symm[0]/evec1_symm[1]:2.6} mal mehr Zombies als Gesunde")
```

Nicht unlogisch: Wenn sich die Zombiefizierung die Waage h√§lt, werden sich auch die tats√§chlichen Zombies und Gesunden langfristig die Waage halten. Der Eigenwert bleibt aber weiterhin "1". Warum √§ndert der sich nicht? Irgendetwas muss an der Matrix also gleich geblieben sein. √úberlegen wir mal, was das sein k√∂nnte. Nun ja, jede Spalte ergibt nachwievor "1", da es sich hier um eine [√úbergangsmatrix](https://de.wikipedia.org/wiki/√úbergangsmatrix) handelt. Die Matrix beh√§lt also bei jeder Multiplikation die Abteilungsgr√∂√üe konstant: Sie wird nicht gr√∂√üer und nicht kleiner.

Nehmen wir jetzt mal an, dass 10% der Zombies komplett drauf gehen. Also bleiben Zombies zu 80% Zombies, 10% werden gesund und 10% zerfallen zu Staub. Wir k√∂nnten das abbilden, indem wir eine dritte Spalte und eine dritte Zeile f√ºr die zu Staub zerfallenen aufnehmen. Aber wir wollen ja nun bewusst schauen, was passiert, wenn Summe einer Spalte nicht mehr "1" ergibt. ;)

```python
#                              heute Zombie  heute gesund
#                                        vv  vv
zombifizierungmatrix_staub =  np.array([[0.8, 0.2],  # <- am n√§chsten Tag Zombie
                                        [0.1, 0.8]]) # <- am n√§chsten Tag gesund
print(zombifizierungmatrix_staub)
(eigvals_staub, eigvectors_staub) = LA.eig(zombifizierungmatrix_staub)
evec1_staub = eigvectors_staub[:, 0]
eval1_staub = eigvals_staub[0]
evec1_staub_relation = evec1_staub[0]/evec1_staub[1]
print(f"Der Eigenvektor ist: {evec1_staub}" )
print(f"Die L√§nge des normierten Eigenvektors ist: {LA.norm(evec1_staub, 2)}")
print(f"Die L√§nge des Eigenvektors ist: {LA.norm(evec1_staub*eval1_staub, 2)}")
print(f"Der Eigenwert ist: {eval1_staub:2.4}")
print(f"Es gibt langfristig {evec1_staub_relation:2.6} mal mehr Zombies als Gesunde")
```

Wir sehen zwei Dinge:
1. Das Verh√§ltnis von Zombies uns Gesunden hat sich ge√§ndert. Das macht ja auch sinn, da wir die Matrix ge√§ndert haben. Relativ zu den Gesunden gibt es jetzt weniger Zombies (weil sie h√§ufiger zu Staub zerfallen). Check!
2. Der Eigenwert ist auf einen Wert <1 gesunken. Das muss also irgenwie aussagen, dass die Norm des eines Vektors nach der Multiplikation mit der Matrix kleiner wird. 

Ich hatte weiter oben geschrieben, dass ein Eigenvektor auch nach der Multiplikation mit der Matrix gleich* bleibt. Wenn da nur diese verflixte Sternchen (*) nicht immer gewesen w√§re (_Terms & Conditions may apply_). Die _eigen_ tliche (pun indendet) Wahrheit ist n√§mlich, dass ein Eigenvektor so definiert ist, dass er "nur" seine Richtung beh√§lt. Er kann l√§nger oder k√ºrzer werden oder auch in die entgegengesetzte Richtung zeigen. Und genau das zeigt der Eigenwert an. Er ist ein Faktor, der den Eigenvektor verl√§ngert (>1), verringert (<1 und >-1) oder umkehrt (<-1). Schauen wir uns das mal mit den altbekannten Grafiken an:

```python
anzahl_tage = 20
at_erster_tag = np.array([1, 99])
ergebnis_hist_staub = zombifizierung(at_erster_tag, zombifizierungmatrix_staub, anzahl_tage)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
create_zombie_plots(ergebnis_hist_staub)
ax2.axline((0,0),(100 * evec1_staub_relation,100), marker = 'o', linestyle = '--')
ax3.axline((0, evec1_staub_relation),(anzahl_tage, evec1_staub_relation), marker = 'o', linestyle = '--')
plt.show()
```

Wir k√∂nnen in den Grafiken nun alles wiedererkennen, was uns der Eigenvektor und der Eigenwert sagt.
1. Die Abteilungsgr√∂√üe wird von Tag zu Tag kleiner.
2. Dabei konvergiert das Verh√§ltnis von Zombies zu Gesunden trotz sinkender Abteilungsgr√∂√üe gegen ein bestimmtes Verh√§ltnis.
3. Und dieses Verh√§ltnis ist 1,41 und ein paar Zerquetschte (genau wurzel(2)).

So, jetzt m√ºssen wir nur noch heraus finden, wie es genau zu dem Eigenwert 0,9414... kommt.
Daf√ºr probieren wir der Vollst√§ndigkeit halber noch einmal das Gleiche wie zuletzt: Wir berechnen einen Vektor, der auf der Linie von Grafik 2 liegt - also das Verh√§ltnis von 1,41... abbildet. Dann stecken wir diesen Vektor in die Matrix und schauen was passiert: 

```python
# Input-Vektor beschreiben
ein_beliebiger_eigenvektor_staub = [100 * evec1_staub_relation, 100] 
print(f"Ein beliebiger Eigenvektor: {ein_beliebiger_eigenvektor_staub}")
print(f"Die Norm des Vektors: {LA.norm(ein_beliebiger_eigenvektor_staub, 2):2.6}")

# Output-Vektor nach Zombifizierung berechnen
neuer_evec = zombifizierungmatrix_staub @ ein_beliebiger_eigenvektor_staub

# Output-Vektor beschreiben
print(f"Der Eigenvektor nach Multiplikation: {neuer_evec}")
print(f"Das Verh√§ltnis der Komponenten des Vektors nach Multiplikation: {neuer_evec[0]/neuer_evec[1]:2.6}")
print(f"Die Norm des Vektors nach Multiplikation: {LA.norm(neuer_evec, 2):2.6}")

# Norm input vs. output vergleichen
print(f"Die L√§nge des Vektors ist nach Multiplikation {LA.norm(neuer_evec)/LA.norm(ein_beliebiger_eigenvektor_staub):2.6}-mal so gro√ü, wie vorher.")
print(f"Zur Erinnerung der Eigenwert: {eval1_staub:2.6}")
```

Yay! Wir sehen hier also noch einmal, dass der Eigenwert "nur" ein Faktor ist, der sagt, um wie viel beliebige Eigenvektoren mit einer Multiplikation gr√∂√üer oder kleiner werden. In unserem Fall: ca. -5,9%. Aber Achtung: Der Eigenwert ist wie der Eigenvektor auch nur eine Eigenschaft der Matrix. Der Eigenwert wei√ü ja nicht mit welchem Vektor sich die Matrix multiplizieren soll. Das bedeutet, dass die -5,9% nur gelten, wenn der Vektor ein Eigenvektor ist. Wenn wir einen ganz anderen Vektor (=zeigt in eine andere Richtung) in die Matrix kippen, dann werden die Eigenwerte in diesem Fall noch etwas kleiner. Aber nichtsdestotrotz zeigt uns der Eigenwert immer, ob das "System" irgendwie "gr√∂√üer" oder "kleiner" wird.

Dann vergleichen wir mal, wie wir eine Matrixmultiplikation in pandas durchf√ºhren w√ºrden. Ich nehme hier pandas, da ein Zugriff √ºber eine echte Datenbank neben der Berechnung im Hauptspeicher auch noch Zeit f√ºr das Einlesen ben√∂tigt. So ein Vegleich w√§re also etwas unfair.

```python
at_erster_tag_df = pd.DataFrame(data=[['z', 1], ['g', 99]],columns=['type', 'count'])
display(at_erster_tag_df)
```

Jetzt generieren wir uns die Zombifizierungsmatrix - allerdings in relationaler schreibweise. Wir ben√∂tigen f√ºr die Werte in der Matrix nur eine Spalte und m√ºssen die m√∂glichen Kombinationen (heute Zombie, morgen gesund etc.) √ºber zwei weitere Spalten abbilden.

```python
zombifizierungmatrix_df = pd.DataFrame(data=zip(['z', 'z', 'g', 'g'], ['z', 'g', 'z', 'g'], [0.9, 0.1, 0.2, 0.8]), columns=['type_from', 'type_to', 'wkt'])
display(zombifizierungmatrix_df)
```

Jetzt joinen wir die zwei Tabellen auf die Ausgangssituation und multiplizieren die Wahrscheinlichkeiten mit H√§ufigkeiten des ersten Tages. Danach summieren wir die dadurch entstandenen H√§ufigkeiten des ersten zweiten Tages unter der Gruppierung "type_to".

Das sql-√Ñquivaltent w√§re also:

`select type_to, sum(wkt) from at_erster_tag_df t1 join zombifizierungmatrix_df 2 on (t1.type = t2.type_to) group by type_to`


```python
at_zweiter_tag_df = at_erster_tag_df.merge(zombifizierungmatrix_df, left_on = 'type', right_on = 'type_from')
at_zweiter_tag_df['count_next'] = at_zweiter_tag_df['count'] * at_zweiter_tag_df['wkt']
at_zweiter_tag_df.groupby(['type_to']).sum(['count_next'])[['count_next']]
```

Wir erhalten die gleichen Ergebnisse wie zuvor. Das ist schon einmal gut. Allerdings bleiben uns die bisher entdeckten Annehmlichkeiten der Linearen Algebra verwehrt: Kein Eigenwert, kein Eigenvektor. Und es gibt sogar noch viel mehr sch√∂ne Dinge, wie z.B. die SVD.
